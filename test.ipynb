```python
import requests
import networkx as nx

resp = requests.get("http://localhost:8080/json/congress-network?chamber=H&tag_id=3068")
result = resp.json()
resp.json()
```

```python
G = nx.Graph()
for edge in result['edges']:
    G.add_edge(edge['source'], edge['target'])

len([x for x in nx.connected_components(G)])
```
